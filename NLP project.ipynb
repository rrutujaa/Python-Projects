```python
!pip install googletrans==4.0.0-rc1
from googletrans import Translator, LANGUAGES
def show_supported_languages():
    print("Supported Languages:")
    for lang_code, lang_name in LANGUAGES.items():
        print(f"{lang_name.capitalize()} ({lang_code})")

show_supported_languages()
def translate_text(text, target_languages):
    translator = Translator()

    # Detect the source language
    detection = translator.detect(text)
    source_lang = detection.lang
    confidence = detection.confidence

    # Check if confidence score is available
    if confidence is not None:
        print(f"\nDetected Language: {LANGUAGES[source_lang].capitalize()} (Confidence: {confidence * 100:.2f}%)")
    else:
        print(f"\nDetected Language: {LANGUAGES[source_lang].capitalize()} (Confidence: Not Available)")

    # Translate into multiple target languages
    for target_lang in target_languages:
        translated = translator.translate(text, dest=target_lang)
        print(f"\nTranslated to {LANGUAGES[target_lang].capitalize()} ({target_lang}):")
        print(translated.text)

def batch_translate_texts(texts, target_languages):
    for i, text in enumerate(texts, 1):
        print(f"\nTranslating sentence {i}:")
        translate_text(text, target_languages)
def main():
    print("\n### Welcome to Multi-language Translator ###")
    show_supported_languages()

    # Take input text from user
    mode = input("\nDo you want to translate a single sentence or multiple? (Type 'single' or 'multiple'): ").strip().lower()

    if mode == 'single':
        text = input("Enter the text to translate: ")
        target_languages = input("Enter target languages (comma separated codes, e.g., 'fr,es,de'): ").split(',')

        translate_text(text, target_languages)

    elif mode == 'multiple':
        num_texts = int(input("How many sentences/paragraphs do you want to translate?: "))
        texts = []
        for i in range(num_texts):
            text = input(f"Enter sentence/paragraph {i + 1}: ")
            texts.append(text)

        target_languages = input("Enter target languages (comma separated codes, e.g., 'fr,es,de'): ").split(',')
        batch_translate_texts(texts, target_languages)

    else:
        print("Invalid input. Please type 'single' or 'multiple'.")

main()
```
